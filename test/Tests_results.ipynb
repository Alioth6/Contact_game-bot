### 28.06.2021

#### Тестовый датасет

data/test_sets/004_human_set.csv

#### Модели поиска ответа:

* sum_model - /src/models/sum_model.py. Word2vec model - обучена на смешанном датасете определений (формальных с википедии и пользовательских)
  * лемматизация слов
  * сложение векторов слов
  * поиск ближайшего среди слов с соответствующим префиксом
  
  
* ens_sum_hyp - объединение двух моделей суммирования и гипонимов/гиперонимов (при построении префиксного дерева был использован обученный на смешанном датасете Word2vec). При обработке определения решается, ответ какой из моделей будет передан на выход. Выбор на основе классификатора определений (https://github.com/Alioth6/contact_game_bot/blob/master/src/models/Classification%20with%20LSTM.ipynb)

#### Метрика

Проверяется точное соответствие ответа загаданному слову при нескольких первых известных буквах в загаданном слове (от 0 до 4).

Финальный скор - процент слов, для которых проверка прошла успешно.

sum_model        | 17.48 | 31.54 | 35.68 | 37.17 | 37.55

### 18.12.2020

#### Тестовый датасет

data/test_sets/004_human_set.csv - несловарный датасет, содержит около 3000 определений. Датасет собран осенью 2020 при работе бота на платформе Heroku.

#### Модели поиска ответа:

* hyp_model - /src/models/hyp_model.py. На базе семантической сети wikiwordnet.
  * лемматизация слов
  * поиск базовой биграммы фразы
  * поиск гипонимов и гиперонимов для слов базовой биграммы


* sum_model - /src/models/sum_model.py. Word2vec model - сжатая версия fasttext модели от RusVectores [ft_freqprune_100K_20K_pq_300.bin](https://github.com/avidale/compress-fasttext/releases/tag/v0.0.1).
  * лемматизация слов
  * сложение векторов слов
  * поиск ближайшего среди слов с соответствующим префиксом
  
  
* sum_plus_hyp_model - объединение двух упомянутых моделей. На выход передаётся до двух слов - результаты от каждой модели. Таким образом мы проверяем, могут ли модели дополнять друг друга - распознают ли они разные определения.


* ens_sum_hyp - объединение двух упомянутых моделей. При обработке определения решается, ответ какой из моделей будет передан на выход. Выбор по числу существительных и глаголов в определении.

#### Метрика

Проверяется точное соответствие ответа загаданному слову (в случае модели sum_plus_hyp_model - наличие загаданного слова в выходном списке).

Финальный скор - процент слов, для которых проверка прошла успешно.

#### Таблица результатов

Модель        | it/s | Score, %
------------- | ----------- | -----
hyp_model        | 86.63 | 7,9
sum_model        | 1.72 | 16.65
sum_plus_hyp_model        | 1.79 | 22
ens_sum_hyp        | 1.38 | 17.6


### 15.05

#### Тестовый датасет

data/test_sets/test_set_human_003.csv - несловарный датасет, содержит около 1000 определений. Слова выбраны случайным образом из около 20 тысяч самых употребительных существительных.

#### Модели поиска ответа:

* hyp_model - /src/models/hyp_model.py. На базе семантической сети wikiwordnet.
  * лемматизация слов
  * поиск базовой биграммы фразы
  * поиск гипонимов и гиперонимов для слов базовой биграммы
  
  
* sim_model - /src/models/sim_model.py. Word2vec model - RDT.
  * поиск 30 близких слов для каждого слова из определения
  * ранжирование по частоте появления в процессе поиска


* sum_model - /src/models/sum_model.py. Word2vec model - RDT.
  * сложение векторов слов из определения
  * поиск 40 ближайших слов
  

RDT - [Russian Distributional Thesaurus](https://nlpub.mipt.ru/Russian_Distributional_Thesaurus), "легкая модель"

#### Метрика

Проверяется, есть ли определяемое слово в топе-3 полученного списка.

Финальный скор - процент слов, для которых проверка прошла успешно.

#### Таблица результатов

------------- | ----------------  | ----------- | -----

### 10.04

#### Модели:
* wiki0 - обучена на датасете из википедии (файл wikipedia_data0.csv, 845 Mb).
  * Модель: skip-gram
  * Размер контекстного окна: 10 слов
  * Размерность векторного пространства: 100
  * Количество итераций: 3
  * Минимальная частота слова в корпусе: 5
  
  
* wiki4 - аналогична wiki0, дообучена на датасете из википедии (файлы wikipedia_data0.csv - wikipedia_data4.csv общим размером 5,5 Gb).


* wiki0_cbow - аналогична wiki0, за исключением:
  * Модель: cbow
  * Размер контекстного окна: 5 слов
  
  
* RDT - [Russian Distributional Thesaurus](https://nlpub.mipt.ru/Russian_Distributional_Thesaurus), наиболее "легкая модель"
  * Модель: skip-gram
  * Размер контекстного окна: 10 слов
  * Размерность векторного пространства: 100
  * Количество итераций: 1
  * Минимальная частота слова в корпусе: 100

#### Тестовый датасет

data/test_sets/test_set_001.csv - датасет составлен из данных викисловаря, содержит 50,5 тысяч определений.

#### Алгоритм поиска ответа

Фраза-определение разбивается на слова, из него удаляются стоп-слова, слова с некириллическими знаками. Для каждого слова из модели выбираются несколько (по умолчанию 10) наиболее близких слов. Функция поиска возвращает список всех найденных слов, ранжированный по частоте их возникновения в процессе поиска.

#### Метрика

Проверяется, есть ли определяемое слово в полученном списке близких слов.

Финальный скор - процент слов, для которых проверка прошла успешно.
